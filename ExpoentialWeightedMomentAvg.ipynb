```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("DailyDelhiClimateTest.csv")
```

```python
df.head()
```

```python
df=df[['date','meantemp']]
```

```python
df.head(4)
```

```python
plt.scatter(df['date'], df['meantemp'])
```

```python
EWMA=df['meantemp'].ewm(alpha=0.9).mean()
```

```python
EWMA
```

```python
df['EWMA']=EWMA
```

```python
df.head()
```

```python
plt.scatter(df['date'], df['meantemp'], color="red")
plt.plot(df['date'], EWMA, color="green")
```
