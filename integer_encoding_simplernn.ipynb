```python
import numpy as np
import pandas as pd

import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer

from keras.datasets import imdb
from keras import Sequential
from keras.layers import Dense,SimpleRNN,Embedding,Flatten
```

**Integer Encoding**

```python
docs = ['go india',
		'india india',
		'hip hip hurray',
		'jeetega bhai jeetega india jeetega',
		'bharat mata ki jai',
		'kohli kohli',
		'sachin sachin',
		'dhoni dhoni',
		'modi ji ki jai',
		'inquilab zindabad']
```

```python
tokenizer=Tokenizer(oov_token='<nothing')
```

```python
tokenizer.fit_on_texts(docs)
```

```python
tokenizer.word_index
```

```python
tokenizer.document_count
```

```python
tokenizer.word_counts
```

```python
sequence=tokenizer.texts_to_sequences(docs)
```

```python
sequence
```

```python
from keras.utils import pad_sequences
```

```python
sequence=pad_sequences(sequence,padding='post')
```

```python
sequence
```

**Using Interger Encode Real Expample**

```python
(X_train, y_train), (X_test, y_test)=imdb.load_data()
```

```python
len(X_train[1])
```

```python
X_train=pad_sequences(X_train, padding='post', maxlen=50)  #50 dur fitst 50 lengt data
X_test=pad_sequences(X_test, padding='post', maxlen=50)
```

```python
X_train[1]
```

```python
model=Sequential()
```

```python
model.add(SimpleRNN(32, input_shape=(50,1),return_sequences=False))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train,epochs=5, validation_data=(X_test,y_test))
```
