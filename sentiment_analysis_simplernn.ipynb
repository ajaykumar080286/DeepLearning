```python
docs = ['go india',
		'india india',
		'hip hip hurray',
		'jeetega bhai jeetega india jeetega',
		'bharat mata ki jai',
		'kohli kohli',
		'sachin sachin',
		'dhoni dhoni',
		'modi ji ki jai',
		'inquilab zindabad']
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts(docs)
```

```python
len(tokenizer.word_index)
```

```python
sequences = tokenizer.texts_to_sequences(docs)
sequences
```

```python
from keras.utils import pad_sequences
sequences = pad_sequences(sequences,padding='post')
sequences
```

```python
import tensorflow

from tensorflow.keras.datasets import imdb
from tensorflow import keras
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,SimpleRNN,Embedding,Flatten
```

```python
model=Sequential()
model.add(Embedding(18,output_dim=2, input_length=5))
model.summary()
```

```python
model.compile('adam','accuracy')
```

```python
pred = model.predict(sequences)
print(pred)
```

```python
(X_train,y_train),(X_test,y_test) = imdb.load_data(num_words=11000)
```

```python
X_train = pad_sequences(X_train,padding='post',maxlen=50)
X_test = pad_sequences(X_test,padding='post',maxlen=50)
X_test.shape
```

```python
model1 = Sequential()
model1.add(Embedding(25000, 2, input_length=50))
model1.add(SimpleRNN(32,return_sequences=False))
model1.add(Dense(1, activation='sigmoid'))

model1.summary()
```

```python
model1.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
history = model1.fit(X_train, y_train,epochs=5,validation_data=(X_test,y_test))
```
