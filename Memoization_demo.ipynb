```python
import time
```

```python
def fib(n):
  if n==1 or n==0:
    return 1
  else:
    return fib(n-1)+fib(n-2)
```

```python
start=time.time()
print(fib(4))
print(time.time()-start)
```

```python
start=time.time()
print(fib(30))
print(time.time()-start)
```
