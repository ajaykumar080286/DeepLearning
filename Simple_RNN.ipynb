```python
import pandas as pd
import numpy as np

import tensorflow as tf
from tensorflow import keras
from keras.layers import SimpleRNN,Dense,Embedding,Flatten
from keras.models import Sequential

from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.utils import pad_sequences
```

```python
docs=["i am ajay",
      "i am from india"]
```

```python
token=Tokenizer()
```

```python
token.fit_on_texts(docs)
```

```python
token.word_index
```

```python
token.document_count
```

```python
sequence=token.texts_to_sequences(docs)
```

```python
sequence
```

```python
len(sequence[1])
```

```python
token.word_index
```

```python
len(token.word_index)
```

```python
sequence=pad_sequences(sequence, padding="post")
```

```python
sequence
```

```python
len(token.word_index)
```

**Embedded Encoding**

```python
model=Sequential()
model.add(Embedding(len(token.word_index)+1,output_dim=2,input_shape=(4,)))
model.add(SimpleRNN(32,return_sequences=False))
model.add(Dense(1,activation="sigmoid"))
```

```python
model.summary()
```

```python
X_train=sequence
```

```python
X_train
```

```python
y_train=np.array([1,0])
```

```python
model.compile(loss='binary_crossentropy', optimizer="adam",metrics=["accuracy"])
```

```python
model.fit(X_train, y_train, epochs=5)
```

**Interger Encode**

```python
model=Sequential()
model.add(SimpleRNN(32,return_sequences=False, input_shape=(4,1)))
model.add(Dense(1,activation="sigmoid"))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer="adam",metrics=["accuracy"])
```

```python
model.fit(X_train, y_train, epochs=5)
```

```python
y_pred=model.predict(sequence)
```
